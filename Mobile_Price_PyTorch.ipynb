```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from sklearn.model_selection import train_test_split
from torch.utils.data import DataLoader , Dataset
from sklearn.preprocessing import StandardScaler , LabelEncoder
```

```python
df=pd.read_csv('train.csv')
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,0:-1],df.iloc[:,-1],test_size=0.4,random_state=42)
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
type(X_train)
```

```python
label=LabelEncoder()
y_train=label.fit_transform(y_train)
y_test=label.transform(y_test)
```

```python
type(y_train)
```

```python
X_train_tensor=torch.from_numpy(X_train).float()
X_test_tensor=torch.from_numpy(X_test).float()
y_train_tensor=torch.from_numpy(y_train).long()
y_test_tensor=torch.from_numpy(y_test).long()
```

```python
type(X_train_tensor)
```

```python
X_train.shape[1]
```

```python
class CustomClass(Dataset):
    def __init__(self, features, labels):
        self.features = features
        self.labels = labels

    def __len__(self):
        return len(self.features)  # Corrected length

    def __getitem__(self, index):
        return self.features[index], self.labels[index]
```

```python
dataset=CustomClass(X_train_tensor,y_train_tensor)
```

```python
dataloader=DataLoader(dataset,batch_size=40,shuffle=True)
```

```python

class MyNN(nn.Module):
    def __init__(self, num_feature, num_classes):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(num_feature, 64),
            nn.ReLU(),
            nn.Linear(64, num_classes),
        )

    def forward(self, feature_data):
        return self.model(feature_data)
```

```python
model = MyNN(X_train.shape[1], len(np.unique(y_train)))
```

```python
epochs=50
learning_rate=1e-4
loss_fn=nn.CrossEntropyLoss()
optimizer=torch.optim.Adam(model.parameters(),lr=learning_rate)
```

```python
for i in range(epochs):
    model.train()
    for batch_data, batch_label in dataloader:
        y_pred = model(batch_data)
        loss = loss_fn(y_pred, batch_label)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    print(f'Epoch: {i + 1}, Loss: {loss.item()}')
```

```python
model.eval()
```

```python
total = 0
correct = 0

with torch.no_grad():
    test_dataset = CustomClass(X_test_tensor, y_test_tensor)
    test_dataloader = DataLoader(test_dataset, batch_size=40, shuffle=False)

    for batch_features, batch_labels in test_dataloader:
        outputs = model(batch_features)
        _, predicted = torch.max(outputs, 1)
        total += batch_labels.size(0)
        correct += (predicted == batch_labels).sum().item()

accuracy = correct / total
print(f'Accuracy: {accuracy:.4f}')
```
